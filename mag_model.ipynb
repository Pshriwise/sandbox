```python
from yt.imods import *
import h5py
```

```python
# If necessary files aren't in the current dir, download them
if not os.path.isfile("magnfield.h5m"):
    from urllib import urlretrieve
    urlretrieve("http://data.pyne.io/magnfield.h5m", "magnfield.h5m")
    
if not os.path.isfile("usrbin22.h5m"):
    # usrbin22.h5m is a large file
    # needs to be downloaded in chunks
    import requests
    r = requests.get("http://data.pyne.io/usrbin22.h5m", stream=True)
    with open("usrbin22.h5m" , 'wb') as file:
        for chunk in r.iter_content(chunk_size=1024):
            if chunk:
                file.write(chunk)
                file.flush()
```

```python
# Load data file
pf = load("usrbin.h5m")
```

```python
# Create the desired slice plot
s = SlicePlot(pf, 'x', ('moab','FluxData'), origin='lower-native')
# Load the facet file
f = h5py.File("magnfield.h5m", "r")
# Get the triangle vertices
coords = f["/tstt/nodes/coordinates"][:]
conn = f["/tstt/elements/Tri3/connectivity"][:]
points = coords[conn-1]
# Annotate slice-triangle intersection contours to the plot
s.annotate_triangle_facets(points, plot_args={"colors": 'black'})
# Alter resolution
s.set_buff_size(1600)
# Change log scale limits
s.set_zlim(('moab','FluxData'), 1e-8, 1e-1)

s.display()
```
